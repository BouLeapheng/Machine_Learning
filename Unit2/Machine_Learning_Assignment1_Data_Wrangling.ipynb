```python
import sklearn  #for marchine learning
import numpy as np # computation
import pandas as pd #data management
```

```python
#1. Load the dataset
dataset = pd.read_csv("https://raw.githubusercontent.com/tareqmahmud/IMDBScraping/master/data/movie.csv")
dataset.head(5)
```

```python
#2. show row and column
dataset.shape
```

```python
#Describe the dataset
dataset.describe()
```

```python
#4. Check the variable(numerical / categorical)
numerical_variable = dataset.columns[dataset.dtypes != "object"]
dataset[numerical_variable].head(5)
```

```python
#categorical value
categorical_variable = dataset.columns[dataset.dtypes == "object"]
dataset[categorical_variable].head(5)
```

```python
#4. Missing value
dataset.isnull().head(5)
```

```python
#specific for numerical value
dataset[numerical_variable].isnull().head(5)
```

```python
#Find out the percentage of missing values
dataset[numerical_variable].isnull().sum()
dataset[numerical_variable].isnull().sum()/len(dataset)
```

```python
#Replace the missing values
dataset.fillna('Sonic')
```

```python
#replace specific column
dataset['Rating'].replace(8.5, 9).head(5)
```

```python
dataset.replace(8.5,9)
```

```python
#Replace more than one value
dataset.replace([8.5,7.5], [9, 9]).head(15)
```

```python
#create the row and add it with the original dataset
#Value of variable , instructions about the variable
row_insert = pd.Series(['Sonic', 2022, 150, 8.2, '81', 500000, '100000'], index = ['Name', 'Year', 'Duration', 'Rating', 'MetaScore', 'Vote', 'Gross'])
dataset.append(row_insert, ignore_index=True)
```

```python
#call specific row
#datasetname.iloc[] = base on index
dataset.iloc[0]
dataset.iloc[1:5]
```
